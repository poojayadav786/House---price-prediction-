In the below code we do.
1. We have split our data into training and testing sets, and now is finally the time to train our algorithm.
2. Then as we know are date column have string type and other as numeric so we replace these strings to 0 and 1.As it does affect boz in house prediction date have no role.
3. we then make new dataframe that in which we drop the id and price of the dataset.
4. we  train the model 70% of data (x_train,y_train),and 30% test data(x_test,y_test).

```python
labels=df['price']
conv_dates=[1 if values==2014 else 0 for values in df.date]
df['date']=conv_dates
train=df.drop(['id','price'],axis=1)
```

```python
from sklearn.cross_validation import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(train,labels,test_size=0.30,random_state=2)
```

At this we train our model.

```python
lreg.fit(x_train,y_train)
```

At this point we test are model


```python
lreg.score(x_test,y_test)
```
